```python
import numpy as np
import pandas as pd
from KNearestNeighbor_weighted import KNearestNeighbors
from joblib import dump, load

data = pd.read_csv("dataset/pasien.csv")
X = data.iloc[:, 1:9].values
y = data.iloc[:, -1].values

data.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

X_train
```

```python
from sklearn.preprocessing import StandardScaler, LabelBinarizer
scaler = StandardScaler()

lb = LabelBinarizer()
# binarizer first column in X_Train
lb.fit(X_train[:, 0])
X_train[:, 0] = lb.transform(X_train[:, 0]).flatten()
X_test[:, 0] = lb.transform(X_test[:, 0]).flatten()

X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)

print(X_train, X_test)
```

```python
from sklearn.metrics import classification_report

def normal_input():
    k = int(input('Enter K value: '))
    knn = KNearestNeighbors(k=k)
    knn.fit(X_train=X_train, y_train=y_train)
    y_pred = knn.predict(X_test)
    print(classification_report(y_test, y_pred))

def weighted_input():
    k = int(input('Enter K value: '))
    knn = KNearestNeighbors(k=k,weighted=True)
    knn.fit(X_train=X_train,y_train=y_train)
    y_pred = knn.predict(X_test)
    print(classification_report(y_test, y_pred))

choice = int(
    input(
        '1. Choose 1 for normal KNN Input\n2. Choose 2 for weighted KNN\n'))
if choice == 1:
    normal_input()
if choice == 2:
    weighted_input()
```
